```python
import tiktoken
import openai
import pickle
import os
import numpy as np
import re
from tqdm import tqdm  # 진행 상황 표시용

# 텍스트를 문장 단위로 나누는 함수
def split_into_sentences(text):
    # 한국어와 영어 문장 구분 패턴
    pattern = r'(?<=[.!?])\s+|(?<=[\.\?\!]["\'])\s+|(?<=다\.)\s+|(?<=요\.)\s+|(?<=니다\.)\s+|(?<=세요\.)\s+'
    sentences = re.split(pattern, text)
    return [s.strip() for s in sentences if s.strip()]

# 토큰 수를 계산하는 함수
def count_tokens(text, model="text-embedding-3-small"):
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)
    return len(tokens)

# 텍스트를 토큰 제한에 맞게 청크로 나누는 함수
def chunk_text(text, max_tokens=8000, model="text-embedding-3-small"):
    sentences = split_into_sentences(text)
    chunks = []
    current_chunk = []
    current_tokens = 0
    
    for sentence in sentences:
        sentence_tokens = count_tokens(sentence, model)
        
        # 문장 자체가 토큰 제한을 초과하는 경우 (드문 경우)
        if sentence_tokens > max_tokens:
            # 이미 모은 청크가 있으면 저장
            if current_chunk:
                chunks.append(" ".join(current_chunk))
                current_chunk = []
                current_tokens = 0
            
            # 긴 문장은 단어 단위로 나누기
            words = sentence.split()
            temp_chunk = []
            temp_tokens = 0
            
            for word in words:
                word_tokens = count_tokens(word + " ", model)
                if temp_tokens + word_tokens <= max_tokens:
                    temp_chunk.append(word)
                    temp_tokens += word_tokens
                else:
                    chunks.append(" ".join(temp_chunk))
                    temp_chunk = [word]
                    temp_tokens = word_tokens
            
            if temp_chunk:
                chunks.append(" ".join(temp_chunk))
        
        # 일반적인 경우: 문장을 현재 청크에 추가할 수 있는지 확인
        elif current_tokens + sentence_tokens <= max_tokens:
            current_chunk.append(sentence)
            current_tokens += sentence_tokens
        else:
            # 현재 청크가 가득 차면 저장하고 새 청크 시작
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentence]
            current_tokens = sentence_tokens
    
    # 마지막 청크 저장
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    
    return chunks

# 개선된 임베딩 함수
def embed_text(text, model="text-embedding-3-small"):
    # 텍스트가 너무 길면 청크로 나누어 처리
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)
    
    # 토큰 수가 제한을 초과하면 청크로 나누어 임베딩
    if len(tokens) > 8000:  # 여유를 두고 8000으로 설정
        chunks = chunk_text(text, max_tokens=8000, model=model)
        
        # 각 청크 임베딩
        chunk_embeddings = []
        for chunk in chunks:
            try:
                response = openai.embeddings.create(
                    model=model,
                    input=chunk
                )
                chunk_embeddings.append(response.data[0].embedding)
            except Exception as e:
                print(f"청크 임베딩 중 오류 발생: {str(e)[:100]}...")
                # 오류 발생 시 빈 청크는 건너뛰기
                continue
        
        # 모든 청크 임베딩의 평균 계산
        if chunk_embeddings:
            return list(np.mean(chunk_embeddings, axis=0))
        else:
            raise ValueError("모든 청크 임베딩에 실패했습니다.")
    
    # 정상적인 경우 그대로 임베딩
    response = openai.embeddings.create(
        model=model,
        input=text
    )
    return response.data[0].embedding

```

```python
import os
import pickle
from tqdm import tqdm
from openai import OpenAI

# 파일 경로 설정
file_path = r"D:\Lecture\coxwave_assignment\final_result.pkl"

# 파일이 존재하는지 확인
if not os.path.exists(file_path):
    print(f"파일이 존재하지 않습니다: {file_path}")

# pickle 파일 불러오기
print("FAQ 데이터 로드 중...")
with open(file_path, 'rb') as f:
    final_result = pickle.load(f)
print(f"데이터 로드 완료. 총 {len(final_result)} 개의 QnA 쌍이 있습니다.")

# 청크 생성
print("청크 생성 중...")
max_tokens_per_chunk = 8192
current_chunk = []
current_tokens = 0
embedded_chunks = {}

# QnA 데이터에 대해 반복
for question, answer in tqdm(final_result.items(), desc="청크 생성"):
    question_tokens = count_tokens(question)
    answer_tokens = count_tokens(answer)
    total_tokens = question_tokens + answer_tokens
    
    # 토큰 수가 8192를 초과하는 경우, 청크를 저장하고 새로 시작
    if current_tokens + total_tokens > max_tokens_per_chunk:
        embedded_chunks[len(embedded_chunks)] = current_chunk
        current_chunk = [(question, answer)]
        current_tokens = total_tokens
    else:
        current_chunk.append((question, answer))
        current_tokens += total_tokens

# 마지막 청크 저장
if current_chunk:
    embedded_chunks[len(embedded_chunks)] = current_chunk

print(f"청크 생성 완료. 총 {len(embedded_chunks)} 개의 청크가 생성되었습니다.")

# 임베딩 수행
print("임베딩 수행 중...")
final_embeddings = {}

# 각 청크별로 처리
for chunk_index, chunk in embedded_chunks.items():
    print(f"청크 {chunk_index+1}/{len(embedded_chunks)} 처리 중...")
    
    # 각 질문/답변 쌍을 개별적으로 처리
    for question, answer in tqdm(chunk, desc=f"청크 {chunk_index+1}"):
        try:
            # 질문 임베딩
            question_embedding = embed_text(question)
            
            # 답변 임베딩
            answer_embedding = embed_text(answer)
            
            # 청크의 각 질문과 답변에 대한 임베딩 저장
            final_embeddings[f"{chunk_index}_{question}"] = {
                "question_embedding": question_embedding,
                "answer_embedding": answer_embedding,
                "question": question,
                "answer": answer
            }
            
        except Exception as e:
            print(f"오류 발생: {question[:50]}... - {str(e)[:100]}...")
            # 오류가 발생하면 계속 진행

# 임베딩 결과 저장
print("임베딩 결과 저장 중...")
output_path = "faq_embeddings_with_vectors.pkl"
with open(output_path, 'wb') as f:
    pickle.dump(final_embeddings, f)

print(f"임베딩된 결과가 {output_path}에 저장되었습니다.")
print(f"총 {len(final_embeddings)} 개의 QnA 쌍이 임베딩되었습니다.")
```

```python
collection_name = 'faq_collection'
try:
    faq_collection = chroma_client.get_collection(name=collection_name)
except:
    faq_collection = chroma_client.create_collection(name=collection_name)
```

```python
# ID, 문서, 임베딩을 Chroma에 맞게 준비
id_list = []
documents = []
embeddings = []

for index, chunk in embedded_chunks.items():
    for question, data in chunk.items():
        id_list.append(f"{index}_{question}")  # 고유 ID 생성
        documents.append(data["question"])  # 질문 텍스트
        embeddings.append(data["question_embedding"])  # 질문 임베딩

# Chroma에 데이터 추가
faq_collection.add(
    ids=id_list,
    documents=documents,
    embeddings=embeddings
)

print("임베딩이 Chroma 컬렉션에 추가되었습니다.")
```
