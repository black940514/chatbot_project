```python
import tiktoken
import re
import numpy as np
import openai
from tqdm import tqdm

# 텍스트를 문장 단위로 나누는 함수
def split_into_sentences(text):
    # 한국어와 영어 문장 구분 패턴
    pattern = r'(?<=[.!?])\s+|(?<=[\.\?\!]["\'])\s+|(?<=다\.)\s+|(?<=요\.)\s+|(?<=니다\.)\s+|(?<=세요\.)\s+'
    sentences = re.split(pattern, text)
    return [s.strip() for s in sentences if s.strip()]

# 토큰 수를 계산하는 함수
def count_tokens(text, model="text-embedding-3-small"):
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)
    return len(tokens)

# QnA 쌍을 오버랩 청킹하는 함수
def chunk_qna_with_overlap(question, answer, max_tokens=8192, overlap_ratio=0.2, model="text-embedding-3-small"):
    """
    질문과 답변을 쌍으로 유지하되, 답변이 너무 길면 오버랩하여 청킹
    
    Parameters:
    - question: 질문 텍스트
    - answer: 답변 텍스트
    - max_tokens: 최대 토큰 수 (기본값: 8192)
    - overlap_ratio: 청크 간 겹치는 비율 (기본값: 0.2)
    - model: 임베딩 모델 이름
    
    Returns:
    - 청크 리스트 [(question, answer_chunk1), (question, answer_chunk2), ...]
    """
    question_tokens = count_tokens(question, model)
    answer_tokens = count_tokens(answer, model)
    
    # 질문과 답변을 합쳐도 최대 토큰 수를 넘지 않으면 그대로 반환
    if question_tokens + answer_tokens <= max_tokens:
        return [(question, answer)]
    
    # 질문 자체가 너무 길면 예외 처리
    if question_tokens > max_tokens // 2:
        raise ValueError(f"질문이 너무 깁니다: {question_tokens} 토큰 (최대 {max_tokens//2})")
    
    # 답변을 문장 단위로 나누기
    answer_sentences = split_into_sentences(answer)
    
    # 각 문장의 토큰 수 계산
    sentence_tokens = [count_tokens(s, model) for s in answer_sentences]
    
    # 청크 생성
    chunks = []
    available_tokens = max_tokens - question_tokens  # 질문을 뺀 나머지 토큰
    
    current_chunk = []
    current_tokens = 0
    
    for i, (sentence, tokens) in enumerate(zip(answer_sentences, sentence_tokens)):
        # 현재 문장을 추가해도 토큰 제한을 넘지 않으면 추가
        if current_tokens + tokens <= available_tokens:
            current_chunk.append(sentence)
            current_tokens += tokens
        else:
            # 현재 청크가 있으면 저장
            if current_chunk:
                chunks.append((question, " ".join(current_chunk)))
            
            # 새 청크 시작
            current_chunk = [sentence]
            current_tokens = tokens
            
            # 문장 자체가 토큰 제한을 초과하면 단어 단위로 나누기
            if tokens > available_tokens:
                words = sentence.split()
                temp_chunk = []
                temp_tokens = 0
                
                for word in words:
                    word_tokens = count_tokens(word + " ", model)
                    if temp_tokens + word_tokens <= available_tokens:
                        temp_chunk.append(word)
                        temp_tokens += word_tokens
                    else:
                        chunks.append((question, " ".join(temp_chunk)))
                        temp_chunk = [word]
                        temp_tokens = word_tokens
                
                if temp_chunk:
                    current_chunk = temp_chunk
                    current_tokens = temp_tokens
                else:
                    current_chunk = []
                    current_tokens = 0
    
    # 마지막 청크 저장
    if current_chunk:
        chunks.append((question, " ".join(current_chunk)))
    
    # 오버랩 적용
    if len(chunks) > 1 and overlap_ratio > 0:
        overlapped_chunks = [chunks[0]]
        
        for i in range(1, len(chunks)):
            prev_answer = chunks[i-1][1]
            curr_answer = chunks[i][1]
            
            # 이전 청크의 문장들
            prev_sentences = split_into_sentences(prev_answer)
            
            # 오버랩할 문장 수 계산 (최소 1개)
            overlap_count = max(1, int(len(prev_sentences) * overlap_ratio))
            
            # 이전 청크의 마지막 문장들 가져오기
            overlap_sentences = prev_sentences[-overlap_count:] if overlap_count < len(prev_sentences) else prev_sentences
            
            # 현재 청크 앞에 오버랩 문장 추가
            overlapped_answer = " ".join(overlap_sentences) + " " + curr_answer
            
            # 오버랩된 청크가 토큰 제한을 초과하면 조정
            while count_tokens(overlapped_answer, model) + question_tokens > max_tokens and overlap_sentences:
                overlap_sentences = overlap_sentences[1:]  # 첫 번째 오버랩 문장 제거
                overlapped_answer = " ".join(overlap_sentences) + " " + curr_answer
            
            overlapped_chunks.append((question, overlapped_answer))
        
        return overlapped_chunks
    
    return chunks

# 딕셔너리 형태의 QnA 데이터를 처리하는 함수
def process_qna_dict(qna_dict, max_tokens=8192, overlap_ratio=0.2, model="text-embedding-3-small"):
    """
    딕셔너리 형태의 QnA 데이터를 오버랩 청킹하여 처리
    
    Parameters:
    - qna_dict: 질문-답변 딕셔너리 {question: answer, ...}
    - max_tokens: 최대 토큰 수
    - overlap_ratio: 청크 간 겹치는 비율
    - model: 임베딩 모델 이름
    
    Returns:
    - 청크 딕셔너리 {chunk_id: {"question": q, "answer": a, "original_q": orig_q}, ...}
    """
    chunked_qna = {}
    chunk_id = 0
    
    for question, answer in tqdm(qna_dict.items(), desc="QnA 청킹"):
        # 각 QnA 쌍을 청킹
        chunks = chunk_qna_with_overlap(question, answer, max_tokens, overlap_ratio, model)
        
        # 청크 저장
        for q, a_chunk in chunks:
            chunked_qna[chunk_id] = {
                "question": q,
                "answer": a_chunk,
                "original_question": question  # 원본 질문 저장 (필요시)
            }
            chunk_id += 1
    
    return chunked_qna

# 임베딩 함수
def embed_chunked_qna(chunked_qna, model="text-embedding-3-small"):
    """
    청킹된 QnA 데이터를 임베딩
    
    Parameters:
    - chunked_qna: 청킹된 QnA 딕셔너리
    - model: 임베딩 모델 이름
    
    Returns:
    - 임베딩이 추가된 딕셔너리
    """
    embedded_qna = {}
    
    for chunk_id, chunk_data in tqdm(chunked_qna.items(), desc="임베딩 생성"):
        question = chunk_data["question"]
        answer = chunk_data["answer"]
        
        try:
            # 질문 임베딩
            q_response = openai.embeddings.create(
                model=model,
                input=question
            )
            question_embedding = q_response.data[0].embedding
            
            # 답변 임베딩
            a_response = openai.embeddings.create(
                model=model,
                input=answer
            )
            answer_embedding = a_response.data[0].embedding
            
            # 임베딩 저장
            embedded_qna[chunk_id] = {
                **chunk_data,
                "question_embedding": question_embedding,
                "answer_embedding": answer_embedding
            }
        except Exception as e:
            print(f"청크 {chunk_id} 임베딩 중 오류 발생: {str(e)[:100]}...")
            continue
    
    return embedded_qna

import pickle
import os

# 파일 경로 설정
file_path = r"D:\Lecture\coxwave_assignment\final_result.pkl"

# 파일이 존재하는지 확인
if not os.path.exists(file_path):
    print(f"파일이 존재하지 않습니다: {file_path}")

# pickle 파일 불러오기
print("FAQ 데이터 로드 중...")
with open(file_path, 'rb') as f:
    qna_dict = pickle.load(f)
print(f"데이터 로드 완료. 총 {len(qna_dict)} 개의 QnA 쌍이 있습니다.")

# QnA 청킹
print("QnA 청킹 중...")
chunked_qna = process_qna_dict(qna_dict, max_tokens=6000, overlap_ratio=0.2)
print(f"청킹 완료. 총 {len(chunked_qna)} 개의 청크가 생성되었습니다.")

# 임베딩 생성
print("임베딩 생성 중...")
embedded_qna = embed_chunked_qna(chunked_qna)
print(f"임베딩 완료. 총 {len(embedded_qna)} 개의 임베딩이 생성되었습니다.")

# 결과 저장
output_path = "faq_embeddings_with_vectors_overlapped.pkl"
with open(output_path, 'wb') as f:
    pickle.dump(embedded_qna, f)
print(f"임베딩된 결과가 {output_path}에 저장되었습니다.")
```

```python
import chromadb

client = chromadb.Client()
client.delete_collection(name="faq_collection")  # 있던 거 초기화 (선택)
faq_collection = client.create_collection(name="faq_collection")
```

```python
id_list = []
documents = []
embeddings = []
metadatas = []

for idx, chunk in embedded_chunks.items():
    id_list.append(f"faq_{idx}")
    
    # ✅ 문서: 질문만 또는 질문+답변 결합 가능
    documents.append(chunk["question"])  # 또는 f"{chunk['question']}\n{chunk['answer']}"
    
    embeddings.append(chunk["question_embedding"])  # ✅ 반드시 384차원
    
    # ✅ 메타데이터 (검색 결과에서 참고용)
    metadatas.append({
        "answer": chunk["answer"],
        "original_question": chunk["original_question"]
    })

# Chroma에 업로드
faq_collection.add(
    ids=id_list,
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas
)

print("✅ Chroma에 384차원 임베딩으로 QnA 등록 완료")
```

## 유사도 검색

```python
query = "미성년자도 판매 회원 등록 가능한가요?"

# 반드시 같은 모델(text-embedding-3-small)로 임베딩
def get_embedding(text: str) -> list[float]:
    response = openai.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    return response.data[0].embedding

query_vec = get_embedding(query)
```

```python
retrived_faq = faq_collection.query(
    query_embeddings=[query_vec],
    n_results=5,
    include=["documents", "metadatas"]
)
```
