```python

import pickle
import os

# 파일 경로 설정
file_path = r"D:\Lecture\coxwave_assignment\final_result.pkl"

# 파일이 존재하는지 확인
if os.path.exists(file_path):
    # pickle 파일 불러오기
    with open(file_path, 'rb') as f:
        final_result = pickle.load(f)
    print(f"파일을 성공적으로 불러왔습니다. 데이터 타입: {type(final_result)}")
else:
    print(f"파일이 존재하지 않습니다: {file_path}")
```

```python
# 1. KoNLPy의 다른 형태소 분석기 활용
from konlpy.tag import Okt, Kkma, Hannanum, Komoran

def extract_keywords_from_qna_konlpy(qna_data, analyzer_type='okt'):
    """
    KoNLPy의 다양한 형태소 분석기를 활용하여 키워드 추출
    
    Parameters:
    - qna_data: 질문-답변 딕셔너리
    - analyzer_type: 사용할 형태소 분석기 ('okt', 'kkma', 'hannanum', 'komoran')
    
    Returns:
    - 질문별 키워드 딕셔너리
    """
    keywords = {}
    
    # 형태소 분석기 선택
    if analyzer_type == 'okt':
        analyzer = Okt()
        extract_func = analyzer.nouns
    elif analyzer_type == 'kkma':
        analyzer = Kkma()
        extract_func = analyzer.nouns
    elif analyzer_type == 'hannanum':
        analyzer = Hannanum()
        extract_func = analyzer.nouns
    else:
        raise ValueError("지원하지 않는 형태소 분석기입니다.")
    
    for question, answer in qna_data.items():
        # 질문에서 명사 추출
        question_keywords = extract_func(question)
        # 답변에서 명사 추출
        answer_keywords = extract_func(answer)
        # 중복 제거 후 질문과 답변에서 명사 합치기
        keywords[question] = list(set(question_keywords + answer_keywords))
        
    return keywords

# 사용 예시
# keywords = extract_keywords_from_qna_konlpy(final_result, 'okt')
```
